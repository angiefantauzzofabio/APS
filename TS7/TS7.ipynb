## Introducción teórica

Un sistema discreto se describe en el dominio temporal mediante una ecuación en diferencias. Las ecuaciones dadas en este trabajo son de la forma:
$$y(n) = \sum_{k=0}^{M} b_k x(n-k)$$

Para analizar la estructura y estabilidad de un sistema, utilizamos la **Transformada Z**. Aplicar esta transformada a una ecuación en diferencias convierte las operaciones de convolución y retardo temporal en operaciones algebraicas en el plano complejo Z.

La función de Transferencia $T(z)$ (o $H(z)$) se define como:
$$T(z) = \frac{Y(z)}{X(z)}$$

Una de las características importantes para el diseño de filtros es su **Respuesta en Frecuencia** $H(j\omega)$, que describe cómo el sistema modifica la amplitud y la fase de las diferentes componentes frecuenciales de una señal de entrada.

Esta respuesta se obtiene evaluando la función de transferencia $T(z)$ sobre el círculo unitario en el plano Z, haciendo la sustitución:
$$z = e^{j\omega}$$
Donde $\omega$ es la frecuencia angular normalizada.

La respuesta en frecuencia es un número complejo y se analiza en términos de:
* **Módulo $|H(j\omega)|$ (Ganancia/Amplitud):** Indica la amplificación o atenuación que sufre cada frecuencia.
* **Fase $\angle H(j\omega)$:** Indica el desplazamiento temporal (retardo) que experimenta cada frecuencia.

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt


b = [1, 1, 1, 1]      # Coeficientes del numerador del filtro
a = [1]               # Coeficientes del denominador del filtro

w, h = signal.freqz(b, a, worN=1000)


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(w, np.abs(h))
plt.title("Módulo")
plt.xticks(np.arange(0, np.pi + 0.01, np.pi / 4), 
           [r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'])
plt.xlabel("ω [rad/muestra]")
plt.ylabel("|T(e^{jω})|")
plt.grid()


plt.subplot(1, 2, 2)
plt.plot(w, np.angle(h))
plt.title("Fase")
plt.xticks(np.arange(0, np.pi + 0.01, np.pi / 4), 
           [r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'])
plt.xlabel("ω [rad/muestra]")
plt.ylabel("∠T(e^{jω}) [rad]")
plt.grid()

plt.tight_layout()
plt.show()
```
